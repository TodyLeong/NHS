```python
import pandas as pd

# Create a new DataFrame with only text values
tweets_text = tweets['tweet_full_text']

# Create an empty list to store hashtags
tags = []

# Extract hashtags from each tweet using a loop
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())

# Create a Pandas Series to count the values in the list
tags_series = pd.Series(tags)

# Display the first 30 records
print(tags_series.value_counts().head(30))
```

```python
# Identify the top trending hashtags with a visualisation.

data = tags_series.value_counts().reset_index()

# Rename the columns to 'word' and 'count'
data.columns = ['word', 'count']

# Ensure the count data type is an integer
data['count'] = data['count'].astype(int)


# Top 10 no of hashtags
top_10_data = data.head(10)

# Sense-check the DataFrame
print(top_10_data.head(10))


# Create a Seaborn barplot of all records with a count > 10
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='word', data=top_10_data[top_10_data['count'] > 10], palette='viridis')
plt.title('Most Common Hashtags')
plt.xlabel('Count')
plt.ylabel('Hashtag')
plt.show()

# Remove overrepresented hashtags
threshold = 100  
filtered_data = top_10_data[top_10_data['count'] <= threshold]

# Sense-check the DataFrame
print(filtered_data.head(10))

# Create a Seaborn barplot of the remaining most frequently used hashtags
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='word', data=filtered_data, palette='viridis')
plt.title('Most Frequently Used Hashtags (Filtered)')
plt.xlabel('Count')
plt.ylabel('Hashtag')
plt.show()
```
